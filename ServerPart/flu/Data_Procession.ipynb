```python
import pandas as pd
from pandas import DataFrame
from sklearn import preprocessing
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

```python
features = pd.read_csv("training_set_features.csv")
labels = pd.read_csv("training_set_labels.csv")
features = features .drop(['employment_industry','employment_occupation','respondent_id','hhs_geo_region',], axis=1)
labels = labels .drop(['h1n1_vaccine','respondent_id'], axis=1)
features.head()
```

```python
#old income_poverty (income_poverty	=[Below Poverty, <75000$, above Poverty, >75000])
```

```python
features_columns = features.columns.tolist()
features_value = features.values.tolist()
labels_columns = labels.columns.tolist()
labels_value = labels.values.tolist()
new_columns = labels_columns + features_columns
new_values = []
for i in range(len(features_value)):
    new_values.append(labels_value[i]+features_value[i])

ndf = DataFrame(new_values,columns=new_columns)
ndf = ndf.dropna()

le1 = preprocessing.LabelEncoder()
le1.fit(ndf['age_group'])
ndf.age_group = le1.transform(ndf['age_group'])
le2 = preprocessing.LabelEncoder()
le2.fit(ndf['race'])
ndf.race = le2.transform(ndf['race'])
le3 = preprocessing.LabelEncoder()
le3.fit(ndf['sex'])
ndf.sex = le3.transform(ndf['sex'])
le4 = preprocessing.LabelEncoder()
le4.fit(ndf['income_poverty'])
ndf.income_poverty = le4.transform(ndf['income_poverty'])
le5 = preprocessing.LabelEncoder() 
le5.fit(ndf['marital_status'])
ndf.marital_status = le5.transform(ndf['marital_status'])
le6 = preprocessing.LabelEncoder()
le6.fit(ndf['rent_or_own'])
ndf.rent_or_own= le6.transform(ndf['rent_or_own'])
le7 = preprocessing.LabelEncoder()
le7.fit(ndf['employment_status'])
ndf.employment_status = le7.transform(ndf['employment_status'])
le8 = preprocessing.LabelEncoder()
le8.fit(ndf['census_msa'])
ndf.census_msa = le8.transform(ndf['census_msa'])
le9 = preprocessing.LabelEncoder()
le9.fit(ndf['education'])
ndf.education = le9.transform(ndf['education'])
train_dataset =ndf.sample(frac=0.7, random_state=1)
test_dataset = ndf.drop(train_dataset.index)
train_dataset.to_csv('train_dataset.csv', encoding='utf-8', index=False)
test_dataset.to_csv('test_dataset.csv', encoding='utf-8', index=False)
```
